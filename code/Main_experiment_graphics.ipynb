```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_acc_disable = np.loadtxt('./disable_train_acc.txt')
train_acc_disable_base = np.loadtxt('./results/disable_train_acc_base.txt')
train_acc_disable_random = np.loadtxt('./results/disable_train_acc_random.txt')
val_acc_disable = np.loadtxt('./disable_val_acc.txt')
val_acc_disable_base = np.loadtxt('./results/disable_val_acc_base.txt')
val_acc_disable_random = np.loadtxt('./results/disable_val_acc_random.txt')
```

```python
(val_acc_disable - val_acc_disable_base).mean()
```

```python
disable_params = np.linspace(0, 1, 51)
```

```python
plt.figure(figsize=(15, 8))
plt.title(' ,   ', fontsize=20)
plt.grid(ls=':')
plt.plot(disable_params * 100, train_acc_disable.mean(axis=1), color='r', label='train acc (precondition)')
plt.plot(disable_params * 100, train_acc_disable_base.mean(axis=1), color='g', label='train acc (min_grad)')
plt.plot(disable_params * 100, train_acc_disable_random.mean(axis=1), color='b', label='train acc (random)')
plt.legend(fontsize=16)
plt.ylim((0.9, 1))
plt.xlabel('  ', fontsize=16)
plt.ylabel('Accuracy', fontsize=16)
```

```python
plt.figure(figsize=(15, 8))
plt.title(' ,   ', fontsize=20)
plt.grid(ls=':')
plt.plot(disable_params * 100, val_acc_disable.mean(axis=1), color='r', label='val acc (precondition)')
plt.plot(disable_params * 100, val_acc_disable_base.mean(axis=1), color='g', label='val acc (min_grad)')
plt.plot(disable_params * 100, val_acc_disable_random.mean(axis=1), color='b', label='val acc (random)')
plt.legend(fontsize=16)
plt.ylim((0.9, 1))
plt.xlabel('  ', fontsize=16)
plt.ylabel('Accuracy', fontsize=16)
```

```python
train_acc_prune = np.loadtxt('./results/prune_train_acc.txt')
train_acc_prune_random = np.loadtxt('./results/prune_train_acc_random.txt')
val_acc_prune = np.loadtxt('./results/prune_val_acc.txt')
val_acc_prune_random = np.loadtxt('./results/prune_val_acc_random.txt')
```

```python
(val_acc_prune - val_acc_prune_random).mean()
```

```python
prune_params = np.linspace(0, 1, 51)[:-1]
```

```python
plt.figure(figsize=(15, 8))
plt.title(' ,   ', fontsize=20)
plt.grid(ls=':')
plt.plot(prune_params * 100, train_acc_prune.mean(axis=1), color='r', label='train acc (min_weights)')
plt.plot(prune_params * 100, train_acc_prune_random.mean(axis=1), color='b', label='train acc (random)')
plt.legend(fontsize=16)
plt.ylim((0.8, 1))
plt.xlabel('  ', fontsize=16)
plt.ylabel('Accuracy', fontsize=16)
```

```python
plt.figure(figsize=(15, 8))
plt.title(' ,   ', fontsize=20)
plt.grid(ls=':')
plt.plot(prune_params * 100, val_acc_prune.mean(axis=1), color='r', label='val acc (min_weights)')
plt.plot(prune_params * 100, val_acc_prune_random.mean(axis=1), color='b', label='val acc (random)')
plt.legend(fontsize=16)
plt.ylim((0.8, 1))
plt.xlabel('  ', fontsize=16)
plt.ylabel('Accuracy', fontsize=16)
```
