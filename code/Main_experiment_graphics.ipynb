```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
results_dir = './mnist_results_100_500/'
```

```python
train_loss_disable = np.loadtxt(results_dir + 'disable_train_acc.txt')
train_loss_disable_base = np.loadtxt(results_dir + 'disable_train_acc_base.txt')
train_loss_disable_random = np.loadtxt(results_dir + 'disable_train_acc_random.txt')
val_loss_disable = np.loadtxt(results_dir + 'disable_val_acc.txt')
val_loss_disable_base = np.loadtxt(results_dir + 'disable_val_acc_base.txt')
val_loss_disable_random = np.loadtxt(results_dir + 'disable_val_acc_random.txt')
```

```python
(val_loss_disable - val_loss_disable_base).mean()
```

```python
disable_params = np.linspace(0, 1, 51)
```

```python
plt.figure(figsize=(15, 8))
plt.title(' ,   ', fontsize=20)
plt.grid(ls=':')
plt.plot(disable_params * 100, train_loss_disable.mean(axis=1), color='r', label='train loss (precondition)')
plt.plot(disable_params * 100, train_loss_disable_base.mean(axis=1), color='g', label='train loss (min_grad)')
plt.plot(disable_params * 100, train_loss_disable_random.mean(axis=1), color='b', label='train loss (random)')
plt.legend(fontsize=16)
plt.ylim((0.9, 1))
plt.xlabel('  ', fontsize=16)
plt.ylabel('MSE', fontsize=16)
```

```python
print('mean result of H strategy:', val_loss_disable.mean())
print('mean result of min_grad strategy:', val_loss_disable_base.mean())
print('mean result of random strategy:', val_loss_disable_random.mean())
```

```python
plt.figure(figsize=(15, 8))
plt.title(' ,   ', fontsize=20)
plt.grid(ls=':')
plt.plot(disable_params * 100, val_loss_disable.mean(axis=1), color='r', label='val loss (precondition)')
plt.plot(disable_params * 100, val_loss_disable_base.mean(axis=1), color='g', label='val loss (min_grad)')
plt.plot(disable_params * 100, val_loss_disable_random.mean(axis=1), color='b', label='val loss (random)')
plt.legend(fontsize=16)
plt.ylim((0.9, 1))
plt.xlabel('  ', fontsize=16)
plt.ylabel('MSE', fontsize=16)
```

```python
train_loss_prune = np.loadtxt(results_dir + 'prune_train_acc.txt')
train_loss_prune_random = np.loadtxt(results_dir + 'prune_train_acc_random.txt')
val_loss_prune = np.loadtxt(results_dir + 'prune_val_acc.txt')
val_loss_prune_random = np.loadtxt(results_dir + 'prune_val_acc_random.txt')
```

```python
(val_loss_prune - val_loss_prune_random).mean()
```

```python
prune_params = np.linspace(0, 1, 51)[:-1]
```

```python
plt.figure(figsize=(15, 8))
plt.title(' ,   ', fontsize=20)
plt.grid(ls=':')
plt.plot(prune_params * 100, train_loss_prune.mean(axis=1), color='r', label='train loss (min_weights)')
plt.plot(prune_params * 100, train_loss_prune_random.mean(axis=1), color='b', label='train loss (random)')
plt.legend(fontsize=16)
plt.ylim((0, 1))
plt.xlabel('  ', fontsize=16)
plt.ylabel('MSE', fontsize=16)
```

```python
plt.figure(figsize=(15, 8))
plt.title(' ,   ', fontsize=20)
plt.grid(ls=':')
plt.plot(prune_params * 100, val_loss_prune.mean(axis=1), color='r', label='val loss (min_weights)')
plt.plot(prune_params * 100, val_loss_prune_random.mean(axis=1), color='b', label='val loss (random)')
plt.legend(fontsize=16)
plt.ylim((0., 1))
plt.xlabel('  ', fontsize=16)
plt.ylabel('MSE', fontsize=16)
```
